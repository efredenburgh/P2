```python
import pandas as pd

data1 = '/content/P2/DATA/Air_Traffic_Passenger_Statistics.csv'
df1 = pd.read_csv(data1)
df1.head()
```

```python
summary = df1.describe()
summary
```

```python
import matplotlib.pyplot as plt

counts = df1['Passenger Count'].value_counts()

plt.hist(df1['Passenger Count'], bins=10, edgecolor='black')

plt.title('Distribution of Passenger Count')
plt.xlabel('Passengers')
plt.ylabel('Frequency')

plt.show()
```

```python
plt.hist(df1['Year'], bins=10, edgecolor='black')

plt.title('Distribution of Year')
plt.xlabel('Year')
plt.ylabel('Flight Frequency')

plt.show()
```
